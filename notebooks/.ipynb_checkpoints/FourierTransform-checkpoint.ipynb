```julia
using ApproxFun, ApproxFunRational, AKNS, LinearAlgebra, SpecialFunctions, Plots,
    SparseArrays, AbstractIterativeSolvers, LaTeXStrings, Printf
```

```julia
f = x -> exp(-x^2)
fhat = k -> sqrt(pi)*exp(-k^2/4)

x = -.1:.001:.1
n = 100
Fhat = ODEFT(f,n,1000)
y = map(x -> abs(Fhat(x)[1]-fhat(x)),x)
str = @sprintf("n = %i",n)
p = plot(x,y,lw=2,yaxis=:log,framestyle=:box,minorgrid=:true,label=latexstring(str),tickfont = font(10),legendfontsize = 10,guidefontsize=14)

for n in [200,400,800,1600]
    Fhat = ODEFT(f,n+1,2000)
    y = map(x -> abs(Fhat(x)[1]-fhat(x)),x)
    str = @sprintf("n = %i",n)
    plot!(x,y,lw=2,yaxis=:log,framestyle=:box,label = latexstring(str),xlabel=L"k",ylabel=L"\mathrm{Error}")
end
display(p)
```

```julia
f = x -> exp(-x^2)
fhat = k -> sqrt(pi)*exp(-k^2/4)

xm = -.1:.001:-.001
xp = 0:.001:.1
n = 100
#Fhat = ODEFT(f,n+1,n+1)
Fhat = ODEFT(f,n+1,n+1)
ym = map(x -> abs(Fhat(x)[1]-fhat(x)),xm)
Fhat = ODEFT(f,n,n)
yp = map(x -> abs(Fhat(x)[1]-fhat(x)),xp)
str = @sprintf("n = %i",n)
p = plot([xm;xp],[ym;yp],lw=2,yaxis=:log,framestyle=:box,minorgrid=:true,label=latexstring(str),tickfont = font(10),legendfontsize = 10,guidefontsize=14)

for n in [200,400,800]
    Fhat = ODEFT(f,n+1,n+1)
    ym = map(x -> abs(Fhat(x)[1]-fhat(x)),xm)
    Fhat = ODEFT(f,n,n)
    yp = map(x -> abs(Fhat(x)[1]-fhat(x)),xp)
    str = @sprintf("n = %i",n)
    plot!([xm;xp],[ym;yp],lw=2,yaxis=:log,framestyle=:box,label = latexstring(str),xlabel=L"k",ylabel=L"\mathrm{Error}")
end
display(p)
```

```julia
FT = FourierTransform(1.0)
f = x -> exp(-x^2)
fhat = k -> sqrt(pi)*exp(-k^2/4)

x = -.1:.001:.1
n = 100
Fhat = FT*Fun(zai(z -> exp(-z^2)), OscLaurent(0.0),n)
y = map(x -> abs(Fhat(x)-fhat(x)),x)
str = @sprintf("n = %i",n)
p = plot(x,y,lw=2,yaxis=:log,framestyle=:box,minorgrid=:true,label=latexstring(str),tickfont = font(10),legendfontsize = 10,guidefontsize=14)

for n in [200,400,800]
    Fhat = FT*Fun(zai(z -> exp(-z^2)), OscLaurent(0.0),n)
    y = map(x -> abs(Fhat(x)[1]-fhat(x)),x)
    str = @sprintf("n = %i",n)
    plot!(x,y,lw=2,yaxis=:log,framestyle=:box,label = latexstring(str),xlabel=L"k",ylabel=L"\mathrm{Error}")
end
display(p)
```

```julia
f = x -> exp(-x^2/2)
fhat = x -> sqrt(2pi)*exp(-x^2/2)

xm = -.1:.001:-.001
xp = 0:.001:.1
n = 100
Fhat = ODEFT_Gaussian(f,n,n+2)
ym = map(x -> abs(Fhat(x+0.1im)[1]-fhat(x+0.1im)),xm)
Fhat = ODEFT_Gaussian(f,n,n+2)
yp = map(x -> abs(Fhat(x+0.1im)[1]-fhat(x+0.1im)),xp)
str = @sprintf("n = %i",n)
p = plot(x,[ym;yp],lw=2,yaxis=:log,framestyle=:box,minorgrid=:true,label=latexstring(str),tickfont = font(10),legendfontsize = 10,guidefontsize=14)

for n in [200,400,800]
    Fhat = ODEFT_Gaussian(f,n,n+2)
    ym = map(x -> abs(Fhat(x)[1]-fhat(x)),xm)
    Fhat = ODEFT_Gaussian(f,n,n+2)
    yp = map(x -> abs(Fhat(x)[1]-fhat(x)),xp)
    str = @sprintf("n = %i",n)
    plot!([xm;xp],[ym;yp],lw=2,yaxis=:log,framestyle=:box,label = latexstring(str),xlabel=L"\mathrm{Re}\,k",ylabel=L"\mathrm{Error}")
end
display(p)
```

```julia
f = x -> 1/(x-1.0-1.0im) - 1/(3*x+1.0im)
fhat2 = k -> 2im*pi*(k == 0.0 ? 2.0/3 : ( k > 0 ? exp(1im*k-k) : 1/3exp(-k/3)) )
fhat = k -> fhat2(-k)

xm = -.2:.001:-.001
xp = 0.001:.001:.2
n = 20
Fhat = ODEFT(f,n+1,n+1)
ym = map(x -> abs(Fhat(x)[1]-fhat(x)),xm)
Fhat = ODEFT(f,n,n)
yp = map(x -> abs(Fhat(x)[1]-fhat(x)),xp)
str = @sprintf("n = %i",n)
p = plot([xm;xp],[ym;yp],lw=2,yaxis=:log,framestyle=:box,minorgrid=:true,label=latexstring(str),tickfont = font(10),legendfontsize = 10,guidefontsize=14)

for n in [40,80,160]
    Fhat = ODEFT(f,n+1,n+1)
    ym = map(x -> abs(Fhat(x)[1]-fhat(x)),xm)
    Fhat = ODEFT(f,n,n)
    yp = map(x -> abs(Fhat(x)[1]-fhat(x)),xp)
    str = @sprintf("n = %i",n)
    plot!([xm;xp],[ym;yp],lw=2,yaxis=:log,framestyle=:box,label = latexstring(str),xlabel=L"k",ylabel=L"\mathrm{Error}")
end
display(p)
```
